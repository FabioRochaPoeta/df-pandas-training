```python
import pandas as pd
import numpy as np
```

```python
df_cars=pd.read_csv('cars.csv')
```

```python
df_cars.columns
```

```python
# transformando em nulos os '?'

df_cars.replace('?', np.NaN, inplace=True)
```

```python
df_cars
```

```python
# checando tipo de price

df_cars.dtypes
```

```python
# transformando type de price (object = string) em float (apenas float aceita Null) 
# e de todas as colunas que tem null, menos número de portas, que é por extenso e não adiantaria

p = 'price'
s = 'stroke'
b = 'bore'
pr = 'peak-rpm'
h = 'horsepower'

df_cars[p] = df_cars[p].astype(float)
df_cars[s] = df_cars[s].astype(float)
df_cars[b] = df_cars[b].astype(float)
df_cars[pr] = df_cars[pr].astype(float)
df_cars[h] = df_cars[h].astype(float)
```

```python
# vendo onde tem nulo e a soma dos nulos em toda a tabela

df_cars.isnull().sum().sort_values(ascending=False)
```

```python
# tratando os nulos do price, substituindo pela média (mean) da marca (make)
# primeiro calculando a média de cada marca:

df_cars.groupby('make')['price'].mean()
```

```python
# listando carros com preço nulo

df_cars[df_cars['price'].isnull()]['make'].unique()
```

```python
## fazendo o for para as 4 colunas que tem nulos e já são float (para substituir o nulo pelas médias) - não consegui

for make in df_cars[df_cars['price'].isnull()]['make'].unique():
    df_cars('make')['price'] = df_cars('make')['price'].mean()
```

```python
# tentando sem for, já que não roda, apenas para o preço inicialmente (essa média não considera as marcas)

avg_price = df_cars['price'].astype(float).mean()
df_cars['price'].replace(np.NaN, avg_price, inplace=True)
```

```python
# vendo qual foi a média de preço

avg_price
```

```python
# checando se saíram os nulos do preço - sim, foi de 4 para 0

df_cars.isnull().sum().sort_values(ascending=False)
```

```python
### Exiba os carros acima da média de preço 

df_cars[(df_cars['price'] >= avg_price)]
```

```python
# para exibir todas as colunas e linhas (são 78 linhas, cada uma com um carro):

pd.set_option("display.max_columns", 500)
pd.set_option("display.max_rows", 500)

df_cars[(df_cars['price'] >= avg_price)]
```

```python
### Crie os seguintes filtros: conversiveis, volkswagen, 4 cilindros, tamanho do motor > 120 - ? 
# (Mudei para tamanho do motor > 50 senão não tem nenhum carro com todas essas características ao mesmo tempo)

df_cars[(df_cars['body-style'] == 'convertible') & (df_cars['make'] == 'volkswagen') & 
(df_cars['num-of-cylinders'] == 'four') & (df_cars['engine-size'] >= 50) ]
```

```python
### Explore o dataframe e traga dúvidas (loc, iloc, rename, drop, dropna, fillna, filtros)

# treinando rename - e voltando pra visualização condensada - renomeei a coluna make
# Rename columns or index labels.
# DataFrame.rename(mapper=None, *, index=None, columns=None, axis=None, copy=None, inplace=False, level=None, errors='ignore')[source]

df_cars.rename(columns={"make": "marca"}, inplace=True)

pd.set_option("display.max_columns", 15)
pd.set_option("display.max_rows", 5)

df_cars

```

```python
# treinando dropna - apagando linhas onde há NA - pq nao apagou linha 0 e 1?
# Remove missing values.
# DataFrame.dropna(*, axis=0, how=_NoDefault.no_default, thresh=_NoDefault.no_default, subset=None, inplace=False, ignore_index=False)

df_cars.dropna(inplace=True)
df_cars
```

```python
# treinando drop - apaguei a coluna Marca
# Drop specified labels from rows or columns.
# DataFrame.drop(labels=None, *, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')[source]

df_cars.drop(columns=['marca'], inplace=True)
df_cars
```

```python
# treinando fillna - substituindo todos os NaN por 'teste'
# Fill NA/NaN values using the specified method.
# DataFrame.fillna(value=None, *, method=None, axis=None, inplace=False, limit=None, downcast=None)[source]

df_cars.fillna('Teste', inplace=True)
df_cars
```

```python
# treinando concat
# Concatenate pandas objects along a particular axis.
# pandas.concat(objs, *, axis=0, join='outer', ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=None)
 
new_row = pd.Series({'a': 3, 'b': 4})
pd.concat([df_cars, new_row.to_frame().T], ignore_index=True)
```

```python
# treinando crosstab
# Compute a simple cross tabulation of two (or more) factors. - Get insights
# pandas.crosstab(index, columns, values=None, rownames=None, colnames=None, aggfunc=None, margins=False, margins_name='All', dropna=True, normalize=False)

pd.crosstab(df_cars['make'], df_cars['symboling'])
```

```python
# treinando get_dummies
# Convert categorical variable into dummy/indicator variables. (0/1)
# pandas.get_dummies(data, prefix=None, prefix_sep='_', dummy_na=False, columns=None, sparse=False, drop_first=False, dtype=None)

pd.get_dummies(df_cars)
```

```python
# pd.cut
# Bin values into discrete intervals. - classes
# pandas.cut(x, bins, right=True, labels=None, retbins=False, precision=3, include_lowest=False, duplicates='raise', ordered=True)

classes_bins = [0, 10000, 20000, 30000, 40000, 50000]
labels = ['E', 'D', 'C', 'B', 'A']

pd.cut(df_cars['price'], bins=classes_bins, labels=labels)
```

```python
# np.linspace (numpy)
# Return evenly spaced numbers over a specified interval.
# numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)

np.linspace.(0, 50000, num=5)

```
